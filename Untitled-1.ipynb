```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import warnings


warnings.filterwarnings('ignore')

print("--- İl Bazlı Deprem Risk Analizi Başlatıldı (MODEL V3")
PROVINCES_FILE = r"C:\Users\EREN\Desktop\deprem\tr.json"
FAULTS_FILE = r"C:\Users\EREN\Desktop\deprem\DFY_GEO_WGS84_2013_s.shp"
BUILDING_FILE = r"C:\Users\EREN\Desktop\deprem\bina_verisi.xlsx"
POPULATION_FILE = r"C:\Users\EREN\Desktop\deprem\ilbazlinüfus.csv"
HISTORICAL_QUAKES_FILE = r"C:\Users\EREN\Desktop\deprem\Earthquakes.xlsx"



def normalize_province_name(name):
    if not isinstance(name, str):
        return name
    
    return name.strip().upper() \
        .replace("İ", "I") \
        .replace("Ş", "S") \
        .replace("Ğ", "G") \
        .replace("Ü", "U") \
        .replace("Ö", "O") \
        .replace("Ç", "C") \
        .replace("-", " ") \
        .replace("KAHRAMANMARAŞ", "KAHRAMANMARAS")

print("\n--- ADIM 1: Kırılganlık (Bina ve Nüfus) Verisi İşleniyor ---")

try:
    df_bina = pd.read_excel(BUILDING_FILE, header=1)
    df_bina = df_bina[df_bina['İl'] != 'TÜRKİYE'].reset_index(drop=True)
    df_bina = df_bina[['İl', 'Toplam', '1980 ve Öncesi', '1981 ve 2000 Arası']]

    for col in ['Toplam', '1980 ve Öncesi', '1981 ve 2000 Arası']:
        df_bina[col] = pd.to_numeric(df_bina[col], errors='coerce')

    df_bina['Eski_Hane_Sayisi'] = df_bina['1980 ve Öncesi'] + df_bina['1981 ve 2000 Arası']
    df_bina['Eski_Hane_Orani'] = (df_bina['Eski_Hane_Sayisi'] / df_bina['Toplam']).fillna(0)
    df_bina['Il_Key'] = df_bina['İl'].apply(normalize_province_name)
    print("Bina stoğu verisi (.xlsx) işlendi.")

    df_nufus_raw = pd.read_csv(POPULATION_FILE, delimiter='|', header=2)
    df_nufus_raw = df_nufus_raw.dropna(how='all').reset_index(drop=True)
    
    pop_cols = [col for col in df_nufus_raw.columns if 'İkamet Edilen İl:' in col]
    df_nufus_raw[pop_cols] = df_nufus_raw[pop_cols].apply(pd.to_numeric, errors='coerce')
    
    total_pop = df_nufus_raw[pop_cols].sum(axis=0)
    total_pop.index = total_pop.index.str.replace('İkamet Edilen İl:', '')
    
    df_nufus = pd.DataFrame({'Il': total_pop.index, 'ToplamNufus': total_pop.values})
    df_nufus['Il_Key'] = df_nufus['Il'].apply(normalize_province_name)
    print("Nüfus verisi işlendi.")

    df_vulnerability = pd.merge(
        df_bina[['Il_Key', 'Eski_Hane_Orani', 'Eski_Hane_Sayisi', 'Toplam']],
        df_nufus[['Il_Key', 'ToplamNufus']],
        on='Il_Key',
        how='left'
    )
    print("Bina ve nüfus verileri birleştirildi.")
    
except Exception as e:
    print(f"HATA (ADIM 1): Kırılganlık verisi işlenirken hata oluştu: {e}")


print("\n--- ADIM 2: Tehlike (Mekansal Veri V2) İşleniyor ---")

try:
    gdf_provinces = gpd.read_file(PROVINCES_FILE)
    gdf_faults = gpd.read_file(FAULTS_FILE)
    print("İl sınırları (tr.json) ve Fay hatları (.shp) başarıyla yüklendi.")

    gdf_provinces_proj = gdf_provinces.to_crs("EPSG:32637")
    gdf_faults_proj = gdf_faults.to_crs("EPSG:32637")
    print("Haritalar metrik koordinat sistemine (EPSG:32637) dönüştürüldü.")

    print("Fay hatlarına 10km'lik tampon bölge (buffer) uygulanıyor...")
    fault_buffer = gdf_faults_proj.buffer(10000)
    fault_buffer_union = fault_buffer.unary_union

    print("İllerin 'Fay Tehlike Alanı' hesaplanıyor (Bu işlem biraz sürebilir)...")
    gdf_provinces_proj['Il_Alani_m2'] = gdf_provinces_proj.geometry.area
    gdf_provinces_proj['Fay_Tehlike_Alani_m2'] = gdf_provinces_proj.geometry.intersection(fault_buffer_union).area
    gdf_provinces_proj['Fay_Tehlike_Orani'] = (gdf_provinces_proj['Fay_Tehlike_Alani_m2'] / gdf_provinces_proj['Il_Alani_m2']).fillna(0)
    print("İl bazlı 'Fay Tehlike Oranı' hesaplandı.")
    
    print("\n--- 2d. Tarihsel Tehlike (M4.0+ Depremler) İşleniyor ---")
    
    df_quakes = pd.read_excel(HISTORICAL_QUAKES_FILE) 
    
    df_quakes = df_quakes.dropna(subset=['Longitude', 'Latitude', 'Magnitude'])
    
    
    gdf_quakes = gpd.GeoDataFrame(
        df_quakes, 
        geometry=gpd.points_from_xy(df_quakes.Longitude, df_quakes.Latitude), 
        crs="EPSG:4326"
    )
    # Deprem haritasını da illerle aynı koordinat sistemine (metrik) taşı
    gdf_quakes_proj = gdf_quakes.to_crs(gdf_provinces_proj.crs)
    print(f"{len(gdf_quakes_proj)} adet M4.0+ deprem noktası haritaya döküldü.")

    # İl sınırları haritasına 'Il_Key' ekle (normalleştirilmiş ad)
    gdf_provinces_proj['Il_Key'] = gdf_provinces_proj['name'].apply(normalize_province_name)
    
    print("Depremler il sınırlarına göre sayılıyor (Spatial Join)...")
    gdf_joined = gpd.sjoin(gdf_provinces_proj, gdf_quakes_proj, how="left", predicate="contains")
    
    # İl bazında depremleri say
    quake_counts = gdf_joined.groupby('Il_Key')['index_right'].count()
    
    # Sayım sonuçlarını ana il haritasına ekle
    gdf_provinces_proj['Tarihsel_Deprem_Sayisi'] = gdf_provinces_proj['Il_Key'].map(quake_counts).fillna(0)
    print("İl bazlı 'Tarihsel Deprem Sayısı' (M4.0+) hesaplandı.")
    print("Bulgaristan gibi sınırlar dışındaki depremler otomatik olarak elendi.")

except Exception as e:
    print(f"HATA (ADIM 2): Mekansal veri işlenirken hata oluştu: {e}")


print("\n--- ADIM 3: Tüm Veriler Birleştiriliyor ve Risk Skoru (V2) Oluşturuluyor ---")

try:
    # --- 3a. Kırılganlık ve Tehlike Verilerini Birleştir ---
    gdf_final = gdf_provinces_proj.merge(
        df_vulnerability,
        on='Il_Key',
        how='left'
    )
    print("Tüm veriler (mekansal + tablo) birleştirildi.")

    # --- 3b. Puanları Normalize Et (Min-Max Scaling) ---
    gdf_final['Puan_Tehlike_FAY'] = (gdf_final['Fay_Tehlike_Orani'] - gdf_final['Fay_Tehlike_Orani'].min()) / \
                                    (gdf_final['Fay_Tehlike_Orani'].max() - gdf_final['Fay_Tehlike_Orani'].min())
    
    gdf_final['Puan_Tehlike_TARIHSEL'] = (gdf_final['Tarihsel_Deprem_Sayisi'] - gdf_final['Tarihsel_Deprem_Sayisi'].min()) / \
                                        (gdf_final['Tarihsel_Deprem_Sayisi'].max() - gdf_final['Tarihsel_Deprem_Sayisi'].min())

    gdf_final['Puan_Kirilganlik'] = (gdf_final['Eski_Hane_Orani'] - gdf_final['Eski_Hane_Orani'].min()) / \
                                    (gdf_final['Eski_Hane_Orani'].max() - gdf_final['Eski_Hane_Orani'].min())

    print("Tüm puanlar (Fay, Tarihsel, Bina) 0-1 arasında normalize edildi.")

    # --- 3c. NİHAİ RİSK SKORUNU (V2) HESAPLA ---
    
    Ağırlık_Fay_Tehlikesi = 0.7
    Ağırlık_Tarihsel_Tehlike = 0.3
    Ağırlık_Toplam_Tehlike = 0.5
    Ağırlık_Kirilganlik = 0.5
    
    gdf_final['Puan_Tehlike_TOPLAM'] = (gdf_final['Puan_Tehlike_FAY'] * Ağırlık_Fay_Tehlikesi) + \
                                       (gdf_final['Puan_Tehlike_TARIHSEL'] * Ağırlık_Tarihsel_Tehlike)
                                    
    gdf_final['NIHAI_RISK_SKORU_01'] = (gdf_final['Puan_Tehlike_TOPLAM'] * Ağırlık_Toplam_Tehlike) + \
                                       (gdf_final['Puan_Kirilganlik'] * Ağırlık_Kirilganlik)
                                    
    gdf_final['RISK_SKORU_10'] = 1 + (gdf_final['NIHAI_RISK_SKORU_01'] * 9)
    gdf_final['RISK_SKORU_10'] = gdf_final['RISK_SKORU_10'].fillna(0)
    
    print("Nihai Risk Skoru (V2 - %70 Fay + %30 Tarihsel) hesaplandı.")

except Exception as e:
    print(f"HATA (ADIM 3): Nihai skor hesaplanırken bir hata oluştu: {e}")


print("\n--- ADIM 4: Sonuçlar ve Çıktı (V2) ---")

try:

    df_sonuc = gdf_final[['name', 'RISK_SKORU_10', 
                        'Puan_Tehlike_TOPLAM', 'Puan_Kirilganlik', 
                        'Puan_Tehlike_FAY', 'Puan_Tehlike_TARIHSEL', 
                        'Tarihsel_Deprem_Sayisi', 'Fay_Tehlike_Orani', 
                        'Eski_Hane_Orani', 'ToplamNufus']]
    df_sonuc = df_sonuc.sort_values(by='RISK_SKORU_10', ascending=False)
    

    print("\n--- EN YÜKSEK RİSKLİ 10 İL (Model V2 Sonucuna Göre) ---")
    
    top_10_listesi = df_sonuc[['name', 'RISK_SKORU_10', 'Puan_Tehlike_TOPLAM', 'Puan_Kirilganlik', 'Puan_Tehlike_FAY', 'Puan_Tehlike_TARIHSEL']].head(10)
    
    print(top_10_listesi.to_string())

    try:
        with open("EN_RISKLI_10_IL_V2.txt", "w", encoding="utf-8") as f:
            f.write("--- EN YÜKSEK RİSKLİ 10 İL (Model V2 Sonucuna Göre) ---\n\n")
            f.write("Risk = %50 Toplam Tehlike + %50 Kırılganlık (Eski Bina)\n")
            f.write("Toplam Tehlike = %70 Puan_Tehlike_FAY + %30 Puan_Tehlike_TARIHSEL\n\n")
            f.write("Puan_Tehlike_FAY: İl alanının faylara yakınlık oranı (0-1 arası)\n")
            f.write("Puan_Tehlike_TARIHSEL: İl sınırları içindeki M4.0+ deprem sayısı (0-1 arası)\n")
            f.write("Puan_Kirilganlik: 2000 öncesi bina oranı (0-1 arası)\n\n")
            f.write(top_10_listesi.to_string())
            
        print("\n>>> En riskli 10 il listesi 'EN_RISKLI_10_IL_V2.txt' dosyasına başarıyla kaydedildi.")
    except Exception as e_txt:
        print(f"!!! HATA: Liste dosyaya kaydedilirken hata oluştu: {e_txt}")

    # --- 4c. Sonuç Haritasını Çizdir ve Kaydet ---
    print("\nRisk haritası (V2) oluşturuluyor ve 'Turkiye_Risk_Haritasi_V2.png' olarak kaydediliyor...")
    
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    gdf_final.plot(
        column='RISK_SKORU_10',
        cmap='Reds',
        linewidth=0.5,
        ax=ax,
        edgecolor='0.7',
        legend=True,
        legend_kwds={'label': "Risk Skoru V2 (1-10 Arası)", 'orientation': "horizontal"}
    )
    ax.set_title('Türkiye İl Bazlı Deprem Risk Analizi (Model V2)\nRisk = %50 (%70 Fay + %30 Tarihsel) + %50 Kırılganlık (Eski Bina)', fontsize=14)
    ax.set_axis_off()
    
    plt.savefig("Turkiye_Risk_Haritasi_V2.png", dpi=300)
    print("Harita (V2) başarıyla kaydedildi.")

    gdf_final_output = gdf_final.to_crs(gdf_provinces.crs)
    output_columns = ['name', 'RISK_SKORU_10', 'Puan_Tehlike_TOPLAM', 'Puan_Kirilganlik', 
                    'Puan_Tehlike_FAY', 'Puan_Tehlike_TARIHSEL', 
                    'Tarihsel_Deprem_Sayisi', 'Fay_Tehlike_Orani', 
                    'Eski_Hane_Orani', 'ToplamNufus', 'geometry']
    gdf_final_output = gdf_final_output[output_columns]
    
    gdf_final_output.to_file("TURKIYE_IL_BAZLI_RISK_ANALIZI_V2.geojson", driver="GeoJSON")
    print("Tüm V2 analiz verisi 'TURKIYE_IL_BAZLI_RISK_ANALIZI_V2.geojson' dosyasına kaydedildi.")

    print("\n--- ANALİZ V2 BAŞARIYLA TAMAMLANDI! ---")

except Exception as e:
    print(f"HATA (ADIM 4): Sonuçlar oluşturulurken bir hata oluştu: {e}")
```

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import warnings

# Gerekli Excel okuma kütüphanesi (openpyxl) zaten kurulu olmalı
# Uyarıları bastır
warnings.filterwarnings('ignore')

print("--- İl Bazlı Deprem Risk Analizi Başlatıldı (MODEL V3 - Etki Alanı ve Büyüklük Ağırlıklı) ---")

# --- DOSYA İSİMLERİ ---
PROVINCES_FILE = r"C:\Users\EREN\Desktop\deprem\tr.json"
FAULTS_FILE = r"C:\Users\EREN\Desktop\deprem\DFY_GEO_WGS84_2013_s.shp"
BUILDING_FILE = r"C:\Users\EREN\Desktop\deprem\bina_verisi.xlsx"
POPULATION_FILE = r"C:\Users\EREN\Desktop\deprem\ilbazlinüfus.csv"
HISTORICAL_QUAKES_FILE = r"C:\Users\EREN\Desktop\deprem\Earthquakes.xlsx" 


# --- İL ADI NORMALLEŞTİRME FONKSİYONU ---
def normalize_province_name(name):
    if not isinstance(name, str):
        return name
    
    return name.strip().upper() \
        .replace("İ", "I") \
        .replace("Ş", "S") \
        .replace("Ğ", "G") \
        .replace("Ü", "U") \
        .replace("Ö", "O") \
        .replace("Ç", "C") \
        .replace("-", " ") \
        .replace("KAHRAMANMARAŞ", "KAHRAMANMARAS")

print("\n--- ADIM 1: Kırılganlık (Bina ve Nüfus) Verisi İşleniyor ---")

try:
    # --- 1a. Bina Stoğu Verisini Yükle ve İşle (Excel olarak) ---
    df_bina = pd.read_excel(BUILDING_FILE, header=1)
    df_bina = df_bina[df_bina['İl'] != 'TÜRKİYE'].reset_index(drop=True)
    df_bina = df_bina[['İl', 'Toplam', '1980 ve Öncesi', '1981 ve 2000 Arası']]

    for col in ['Toplam', '1980 ve Öncesi', '1981 ve 2000 Arası']:
        df_bina[col] = pd.to_numeric(df_bina[col], errors='coerce')

    df_bina['Eski_Hane_Sayisi'] = df_bina['1980 ve Öncesi'] + df_bina['1981 ve 2000 Arası']
    df_bina['Eski_Hane_Orani'] = (df_bina['Eski_Hane_Sayisi'] / df_bina['Toplam']).fillna(0)
    df_bina['Il_Key'] = df_bina['İl'].apply(normalize_province_name)
    print("Bina stoğu verisi (.xlsx) işlendi.")

    # --- 1b. Nüfus Verisini Yükle ve İşle ---
    df_nufus_raw = pd.read_csv(POPULATION_FILE, delimiter='|', header=2)
    df_nufus_raw = df_nufus_raw.dropna(how='all').reset_index(drop=True)
    
    pop_cols = [col for col in df_nufus_raw.columns if 'İkamet Edilen İl:' in col]
    df_nufus_raw[pop_cols] = df_nufus_raw[pop_cols].apply(pd.to_numeric, errors='coerce')
    
    total_pop = df_nufus_raw[pop_cols].sum(axis=0)
    total_pop.index = total_pop.index.str.replace('İkamet Edilen İl:', '')
    
    df_nufus = pd.DataFrame({'Il': total_pop.index, 'ToplamNufus': total_pop.values})
    df_nufus['Il_Key'] = df_nufus['Il'].apply(normalize_province_name)
    print("Nüfus verisi işlendi.")

    # --- 1c. Kırılganlık Verilerini Birleştir ---
    df_vulnerability = pd.merge(
        df_bina[['Il_Key', 'Eski_Hane_Orani', 'Eski_Hane_Sayisi', 'Toplam']],
        df_nufus[['Il_Key', 'ToplamNufus']],
        on='Il_Key',
        how='left'
    )
    print("Bina ve nüfus verileri birleştirildi.")
    
except Exception as e:
    print(f"HATA (ADIM 1): Kırılganlık verisi işlenirken hata oluştu: {e}")


print("\n--- ADIM 2: Tehlike (Mekansal Veri V3) İşleniyor ---")

try:
    # --- 2a. Haritaları Yükle ---
    gdf_provinces = gpd.read_file(PROVINCES_FILE)
    gdf_faults = gpd.read_file(FAULTS_FILE)
    print("İl sınırları (tr.json) ve Fay hatları (.shp) başarıyla yüklendi.")

    # --- 2b. Koordinat Sistemini (CRS) Değiştir ---
    gdf_provinces_proj = gdf_provinces.to_crs("EPSG:32637")
    gdf_faults_proj = gdf_faults.to_crs("EPSG:32637")
    print("Haritalar metrik koordinat sistemine (EPSG:32637) dönüştürüldü.")

    # --- 2c. POTANSİYEL TEHLİKE PUANI (FAY YAKINLIĞI) HESAPLAMA ---
    print("Fay hatlarına 10km'lik tampon bölge (buffer) uygulanıyor...")
    fault_buffer = gdf_faults_proj.buffer(10000) # 10,000 metre = 10km
    fault_buffer_union = fault_buffer.unary_union

    print("İllerin 'Fay Tehlike Alanı' hesaplanıyor...")
    gdf_provinces_proj['Il_Alani_m2'] = gdf_provinces_proj.geometry.area
    gdf_provinces_proj['Fay_Tehlike_Alani_m2'] = gdf_provinces_proj.geometry.intersection(fault_buffer_union).area
    gdf_provinces_proj['Fay_Tehlike_Orani'] = (gdf_provinces_proj['Fay_Tehlike_Alani_m2'] / gdf_provinces_proj['Il_Alani_m2']).fillna(0)
    print("İl bazlı 'Fay Tehlike Oranı' hesaplandı.")
    
    # --- 2d. YENİ (V3): TARİHSEL TEHLİKE (BÜYÜKLÜK AĞIRLIKLI) HESAPLAMA ---
    print("\n--- 2d. Tarihsel Tehlike (M4.0+ Depremler) İşleniyor (V3) ---")
    
    df_quakes = pd.read_excel(HISTORICAL_QUAKES_FILE) 
    df_quakes = df_quakes.dropna(subset=['Longitude', 'Latitude', 'Magnitude'])
    # Büyüklük sütununun sayısal olduğundan emin ol
    df_quakes['Magnitude'] = pd.to_numeric(df_quakes['Magnitude'], errors='coerce')
    
    gdf_quakes = gpd.GeoDataFrame(
        df_quakes, 
        geometry=gpd.points_from_xy(df_quakes.Longitude, df_quakes.Latitude), 
        crs="EPSG:4326"
    )
    gdf_quakes_proj = gdf_quakes.to_crs(gdf_provinces_proj.crs)
    print(f"{len(gdf_quakes_proj)} adet M4.0+ deprem noktası haritaya döküldü.")

    # İl sınırları haritasına 'Il_Key' ekle
    gdf_provinces_proj['Il_Key'] = gdf_provinces_proj['name'].apply(normalize_province_name)
    
    # YENİ (V3): İllere 50km 'Etki Alanı' (Buffer) uygula
    print("İllere 50km'lik 'etki alanı' (buffer) uygulanıyor (Ege Denizi vb. sorunları için)...")
    gdf_provinces_proj['etki_alani_50km'] = gdf_provinces_proj.geometry.buffer(50000) # 50,000 metre = 50km
    
    # Analiz için geometriyi bu etki alanına ayarla
    gdf_provinces_buffered = gdf_provinces_proj.set_geometry('etki_alani_50km')
    
    print("Depremler illerin 50km etki alanına göre birleştiriliyor (Spatial Join)...")
    # 'predicate="intersects"' (kesişen) kullanarak 50km alan içindeki depremleri bul
    gdf_joined = gpd.sjoin(gdf_provinces_buffered, gdf_quakes_proj, how="left", predicate="intersects")
    
    # YENİ (V3): Depremleri 'saymak' yerine Magnitüdlerini 'topla'
    print("İl bazlı 'Tarihsel Ağırlıklı Puan' (Magnitüd Toplamı) hesaplanıyor...")
    quake_magnitude_sum = gdf_joined.groupby('Il_Key')['Magnitude'].sum()
    
    # Sayım sonuçlarını *orijinal* il haritasına ekle
    gdf_provinces_proj['Tarihsel_Puan_Toplam_Mag'] = gdf_provinces_proj['Il_Key'].map(quake_magnitude_sum).fillna(0)
    
    print("İl bazlı 'Tarihsel Ağırlıklı Puan' (M4.0+) hesaplandı.")
    print("Ege Denizi ve komşu ülkelerdeki depremler artık kıyı illerini etkiliyor.")

except Exception as e:
    print(f"HATA (ADIM 2): Mekansal veri işlenirken hata oluştu: {e}")


print("\n--- ADIM 3: Tüm Veriler Birleştiriliyor ve Risk Skoru (V3) Oluşturuluyor ---")

try:
    # --- 3a. Kırılganlık ve Tehlike Verilerini Birleştir ---
    # Orijinal geometriye geri dön (etki alanını bırak)
    gdf_final = gdf_provinces_proj.set_geometry('geometry')
    
    gdf_final = gdf_final.merge(
        df_vulnerability,
        on='Il_Key',
        how='left'
    )
    print("Tüm veriler (mekansal + tablo) birleştirildi.")

    # --- 3b. Puanları Normalize Et (Min-Max Scaling) ---
    # 1. Potansiyel Tehlike Puanı (Fay)
    gdf_final['Puan_Tehlike_FAY'] = (gdf_final['Fay_Tehlike_Orani'] - gdf_final['Fay_Tehlike_Orani'].min()) / \
                                    (gdf_final['Fay_Tehlike_Orani'].max() - gdf_final['Fay_Tehlike_Orani'].min())
    
    # 2. Tarihsel Tehlike Puanı (Büyüklük Toplamı) - YENİ (V3)
    gdf_final['Puan_Tehlike_TARIHSEL'] = (gdf_final['Tarihsel_Puan_Toplam_Mag'] - gdf_final['Tarihsel_Puan_Toplam_Mag'].min()) / \
                                        (gdf_final['Tarihsel_Puan_Toplam_Mag'].max() - gdf_final['Tarihsel_Puan_Toplam_Mag'].min())

    # 3. Kırılganlık Puanı (Bina)
    gdf_final['Puan_Kirilganlik'] = (gdf_final['Eski_Hane_Orani'] - gdf_final['Eski_Hane_Orani'].min()) / \
                                    (gdf_final['Eski_Hane_Orani'].max() - gdf_final['Eski_Hane_Orani'].min())

    print("Tüm puanlar (Fay, Tarihsel-Ağırlıklı, Bina) 0-1 arasında normalize edildi.")

    # --- 3c. NİHAİ RİSK SKORUNU (V3) HESAPLA ---
    
    Ağırlık_Fay_Tehlikesi = 0.7
    Ağırlık_Tarihsel_Tehlike = 0.3
    Ağırlık_Toplam_Tehlike = 0.5
    Ağırlık_Kirilganlik = 0.5
    
    gdf_final['Puan_Tehlike_TOPLAM'] = (gdf_final['Puan_Tehlike_FAY'] * Ağırlık_Fay_Tehlikesi) + \
                                       (gdf_final['Puan_Tehlike_TARIHSEL'] * Ağırlık_Tarihsel_Tehlike)
                                    
    gdf_final['NIHAI_RISK_SKORU_01'] = (gdf_final['Puan_Tehlike_TOPLAM'] * Ağırlık_Toplam_Tehlike) + \
                                       (gdf_final['Puan_Kirilganlik'] * Ağırlık_Kirilganlik)
                                    
    gdf_final['RISK_SKORU_10'] = 1 + (gdf_final['NIHAI_RISK_SKORU_01'] * 9)
    gdf_final['RISK_SKORU_10'] = gdf_final['RISK_SKORU_10'].fillna(0)
    
    print("Nihai Risk Skoru (V3 - Etki Alanı + Büyüklük Ağırlıklı) hesaplandı.")

except Exception as e:
    print(f"HATA (ADIM 3): Nihai skor hesaplanırken bir hata oluştu: {e}")


print("\n--- ADIM 4: Sonuçlar ve Çıktı (V3) ---")

try:
    # --- 4a. En Riskli İlleri Sırala ---
    df_sonuc = gdf_final[['name', 'RISK_SKORU_10', 
                        'Puan_Tehlike_TOPLAM', 'Puan_Kirilganlik', 
                        'Puan_Tehlike_FAY', 'Puan_Tehlike_TARIHSEL', 
                        'Tarihsel_Puan_Toplam_Mag', 'Fay_Tehlike_Orani', 
                        'Eski_Hane_Orani', 'ToplamNufus']]
    df_sonuc = df_sonuc.sort_values(by='RISK_SKORU_10', ascending=False)
    
    # --- 4b. En Riskli 10 İli Göster ve Dosyaya Kaydet ---
    print("\n--- EN YÜKSEK RİSKLİ 10 İL (Model V3 Sonucuna Göre) ---")
    
    top_10_listesi = df_sonuc[['name', 'RISK_SKORU_10', 'Puan_Tehlike_TOPLAM', 'Puan_Kirilganlik', 'Puan_Tehlike_FAY', 'Puan_Tehlike_TARIHSEL']].head(10)
    
    print(top_10_listesi.to_string())

    # --- Listeyi .txt dosyasına kaydet ---
    try:
        with open("EN_RISKLI_10_IL_V3.txt", "w", encoding="utf-8") as f:
            f.write("--- EN YÜKSEK RİSKLİ 10 İL (Model V3 Sonucuna Göre) ---\n\n")
            f.write("Risk = %50 Toplam Tehlike + %50 Kırılganlık (Eski Bina)\n")
            f.write("Toplam Tehlike = %70 Puan_Tehlike_FAY + %30 Puan_Tehlike_TARIHSEL\n\n")
            f.write("Puan_Tehlike_FAY: İl alanının faylara yakınlık oranı (0-1 arası)\n")
            f.write("Puan_Tehlike_TARIHSEL: İl sınırlarının 50km etki alanındaki M4.0+ depremlerin BÜYÜKLÜK TOPLAMI (0-1 arası)\n")
            f.write("Puan_Kirilganlik: 2000 öncesi bina oranı (0-1 arası)\n\n")
            f.write(top_10_listesi.to_string())
            
        print("\n>>> En riskli 10 il listesi 'EN_RISKLI_10_IL_V3.txt' dosyasına başarıyla kaydedildi.")
    except Exception as e_txt:
        print(f"!!! HATA: Liste dosyaya kaydedilirken hata oluştu: {e_txt}")

    # --- 4c. Sonuç Haritasını Çizdir ve Kaydet ---
    print("\nRisk haritası (V3) oluşturuluyor ve 'Turkiye_Risk_Haritasi_V3.png' olarak kaydediliyor...")
    
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    gdf_final.plot(
        column='RISK_SKORU_10',
        cmap='Reds',
        linewidth=0.5,
        ax=ax,
        edgecolor='0.7',
        legend=True,
        legend_kwds={'label': "Risk Skoru V3 (1-10 Arası)", 'orientation': "horizontal"}
    )
    ax.set_title('Türkiye İl Bazlı Deprem Risk Analizi (Model V3)\nRisk = %50 (%70 Fay + %30 Tarihsel Büyüklük Top.) + %50 Kırılganlık (Eski Bina)', fontsize=14)
    ax.set_axis_off()
    
    plt.savefig("Turkiye_Risk_Haritasi_V3.png", dpi=300)
    print("Harita (V3) başarıyla kaydedildi.")

    # --- 4d. Sonuç Verisini GeoJSON Olarak Kaydet ---
    gdf_final_output = gdf_final.to_crs(gdf_provinces.crs)
    output_columns = ['name', 'RISK_SKORU_10', 'Puan_Tehlike_TOPLAM', 'Puan_Kirilganlik', 
                    'Puan_Tehlike_FAY', 'Puan_Tehlike_TARIHSEL', 
                    'Tarihsel_Puan_Toplam_Mag', 'Fay_Tehlike_Orani', 
                    'Eski_Hane_Orani', 'ToplamNufus', 'geometry']
    gdf_final_output = gdf_final_output[output_columns]
    
    gdf_final_output.to_file("TURKIYE_IL_BAZLI_RISK_ANALIZI_V3.geojson", driver="GeoJSON")
    print("Tüm V3 analiz verisi 'TURKIYE_IL_BAZLI_RISK_ANALIZI_V3.geojson' dosyasına kaydedildi.")

    print("\n--- ANALİZ V3 BAŞARIYLA TAMAMLANDI! ---")

except Exception as e:
    print(f"HATA (ADIM 4): Sonuçlar oluşturulurken bir hata oluştu: {e}")
```

```python
import pandas as pd
import geopandas as gpd
import warnings
import sys

# Excel okumak için
try:
    import openpyxl
except ImportError:
    print(">>> 'openpyxl' kütüphanesi bulunamadı. Yükleniyor...")
    # pip'i Python içinden çalıştırma (normalde tavsiye edilmez ama debug için OK)
    import subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "openpyxl"])
    print(">>> 'openpyxl' yüklendi. Koda devam ediliyor.")

# Uyarıları bastır
warnings.filterwarnings('ignore')

print("--- Birleştirme (Merge) Hata Ayıklama Testi Başlatıldı ---")

# --- DOSYA İSİMLERİ ---
PROVINCES_FILE = r"C:\Users\EREN\Desktop\deprem\tr.json"
BUILDING_FILE = r"C:\Users\EREN\Desktop\deprem\bina_verisi.xlsx"

# --- İL ADI NORMALLEŞTİRME FONKSİYONU ---
def normalize_province_name(name):
    if not isinstance(name, str):
        return name
    
    # K.Maraş gibi kısaltmaları da düzeltmek için eklemeler yapalım
    name_tr = name.strip().upper() \
        .replace("İ", "I") \
        .replace("Ş", "S") \
        .replace("Ğ", "G") \
        .replace("Ü", "U") \
        .replace("Ö", "O") \
        .replace("Ç", "C") \
        .replace("-", " ") 
    
    # Kısaltma kontrolleri
    if name_tr == "K.MARAS":
        return "KAHRAMANMARAS"
    if name_tr == "K MARAS":
        return "KAHRAMANMARAS"
    if "KAHRAMANMARA" in name_tr: # Kahramanmaraş'ı da kapsar
        return "KAHRAMANMARAS"
    
    # Diğer iller için standartlaştırma
    return name_tr

print("\n--- ADIM 1: Harita Dosyası (tr.json) Okunuyor ---")
try:
    gdf_provinces = gpd.read_file(PROVINCES_FILE)
    gdf_provinces['Il_Key'] = gdf_provinces['name'].apply(normalize_province_name)
    map_keys = set(gdf_provinces['Il_Key'].unique())
    print(f"Harita dosyasında {len(map_keys)} adet il bulundu (Örn: {list(map_keys)[:3]})")
except Exception as e:
    print(f"HATA: Harita dosyası okunamadı: {e}")
    map_keys = set()

print("\n--- ADIM 2: Bina Verisi Dosyası (bina_verisi.xlsx) Okunuyor ---")
try:
    df_bina = pd.read_excel(BUILDING_FILE, header=1)
    # 'İl' sütununun varlığından emin ol
    if 'İl' not in df_bina.columns:
        print("HATA: 'bina_verisi.xlsx' dosyasında 'İl' adında bir sütun bulunamadı!")
        df_bina = None
    else:
        df_bina = df_bina.dropna(subset=['İl']) # İl adı boş olanları at
        df_bina['Il_Key'] = df_bina['İl'].apply(normalize_province_name)
        building_keys = set(df_bina['Il_Key'].unique())
        print(f"Bina dosyasında {len(building_keys)} adet il bulundu (Örn: {list(building_keys)[:3]})")
        
        print("\n--- 'bina_verisi.xlsx' içindeki 'KAHRAMANMARAS' aranıyor ---")
        maraş_check = [key for key in building_keys if "KAHRAMANMARAS" in key]
        if maraş_check:
            print(f"BULUNDU! Bina verisinde 'Kahramanmaraş' şu anahtarla bulundu: {maraş_check}")
        else:
            print("BULUNAMADI! Bina verisinde 'Kahramanmaraş' (veya K.Maraş) bulunamadı veya normalize edilemedi.")

except Exception as e:
    print(f"HATA: Bina dosyası okunamadı: {e}")
    building_keys = set()

print("\n" + "="*50)
print("--- TEST SONUCU ---")

if map_keys and building_keys:
    # --- Esas Test: Haritada olup Bina Verisinde Olmayanlar ---
    map_only = map_keys - building_keys
    
    if not map_only:
        print("\nTEBRİKLER! Haritadaki tüm iller (81) bina verisi dosyasında bulundu.")
        print("Birleştirme hatası (NaN) yok. Sorun başka bir yerde olmalı.")
    else:
        print("\n!!! KRİTİK HATA BULUNDU !!!")
        print("Aşağıdaki iller 'tr.json' (harita) dosyasında var,")
        print("ancak 'bina_verisi.xlsx' dosyasında BULUNAMADI (veya YAZIMI EŞLEŞMEDİ):")
        print("-" * 30)
        for i, il in enumerate(sorted(list(map_only))):
            print(f"{i+1}. {il}")
        print("-" * 30)
        print("\nBu iller, bina verisi olmadığı için haritada 'bembeyaz' (0 puan) çıkıyor.")
        print("Çözüm: 'bina_verisi.xlsx' dosyasındaki bu illerin adını, haritadaki adla (veya 'Kahramanmaraş' gibi) eşleşecek şekilde düzeltin.")

else:
    print("\nTest tamamlanamadı. Dosyalardan biri okunamadığı için karşılaştırma yapılamadı.")
```
